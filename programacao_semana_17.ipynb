```python
# - Código para salvar um texto extraído da internet
import requests
from bs4 import BeautifulSoup

url = "https://sites.google.com/view/vestibular123"  # Substitua pelo URL correto
response = requests.get(url)
content = response.content

soup = BeautifulSoup(content, "html.parser")
meta_tags = soup.find_all('meta', {'property': 'og:description'})

# Extrair o conteúdo da primeira meta tag encontrada
if meta_tags:
    texto_salvar = meta_tags[0]['content']
    print(texto_salvar)
else:
    print("Meta tag 'og:description' não encontrada.")
```
